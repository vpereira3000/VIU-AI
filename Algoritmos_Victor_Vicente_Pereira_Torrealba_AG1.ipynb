##Devolución de cambio por técnica voraz

```
def cambio_monedas(N, SM):
  SOLUCION = [0]*len(SM)   #SOLUCION = [0,0,0,0,..]
  ValorAcumulado = 0

  for i,valor in enumerate(SM):
    monedas =  (N-ValorAcumulado)//valor
    SOLUCION[i] = monedas
    ValorAcumulado = ValorAcumulado + monedas*valor

    if ValorAcumulado == N:
      return SOLUCION


cambio_monedas(15,[25,10,5,1])
```

## N-Reinas por técnica de vueta atrás


```
def escribe(S):
  n = len(S)
  for x in range(n):
    print("")
    for i in range(n):
      if S[i] == x+1:
        print(" X " , end="")
      else:
        print(" - ", end="")


def es_prometedora(SOLUCION,etapa):
  #print(SOLUCION)
  #Si la solución tiene dos valores iguales no es valida => Dos reinas en la misma fila
  for i in range(etapa+1):
    #print("El valor " + str(SOLUCION[i]) + " está " +  str(SOLUCION.count(SOLUCION[i])) + " veces")
    if SOLUCION.count(SOLUCION[i]) > 1:
      return False

    #Verifica las diagonales
    for j in range(i+1, etapa +1 ):
      #print("Comprobando diagonal de " + str(i) + " y " + str(j))
      if abs(i-j) == abs(SOLUCION[i]-SOLUCION[j]) : return False
  return True



def reinas(N, solucion=[], etapa=0):
  if len(solucion) == 0:
      solucion=[0 for i in range(N)]

  for i in range(1, N+1):
    solucion[etapa] = i

    if es_prometedora(solucion, etapa):
      if etapa == N-1:
        print(solucion)
        #escribe(solucion)
        print()
      else:
        reinas(N, solucion, etapa+1)
    else:
      None

    solucion[etapa] = 0

reinas(8)
```

## Viaje por el rio. Programación dinámica

```
TARIFAS = [
[0,5,4,3,999,999,999],
[999,0,999,2,3,999,11],
[999,999, 0,1,999,4,10],
[999,999,999, 0,5,6,9],
[999,999, 999,999,0,999,4],
[999,999, 999,999,999,0,3],
[999,999,999,999,999,999,0]
]



################################################################
def Precios(TARIFAS):
################################################################
  #Total de Nodos
  N = len(TARIFAS[0])

  #Inicialización de la tabla de precios
  PRECIOS = [ [9999]*N for i in [9999]*N]
  RUTA = [ [""]*N for i in [""]*N]

  for i in range(0,N-1):
    RUTA[i][i] = i             #Para ir de i a i se "pasa por i"
    PRECIOS[i][i] = 0          #Para ir de i a i se se paga 0
    for j in range(i+1, N):
      MIN = TARIFAS[i][j]
      RUTA[i][j] = i

      for k in range(i, j):
        if PRECIOS[i][k] + TARIFAS[k][j] < MIN:
            MIN = min(MIN, PRECIOS[i][k] + TARIFAS[k][j] )
            RUTA[i][j] = k          #Anota que para ir de i a j hay que pasar por k
        PRECIOS[i][j] = MIN

  return PRECIOS,RUTA
################################################################


PRECIOS,RUTA = Precios(TARIFAS)
#print(PRECIOS[0][6])

print("PRECIOS")
for i in range(len(TARIFAS)):
  print(PRECIOS[i])

print("\nRUTA")
for i in range(len(TARIFAS)):
  print(RUTA[i])

#Determinar la ruta con Recursividad
def calcular_ruta(RUTA, desde, hasta):
  if desde == hasta:
    #print("Ir a :" + str(desde))
    return ""
  else:
    return str(calcular_ruta( RUTA, desde, RUTA[desde][hasta])) +  \
                ',' + \
                str(RUTA[desde][hasta] \
              )

print("\nLa ruta es:")
calcular_ruta(RUTA, 0,6)
```

```
#Problema: Econtrar los dos puntos mas cercanos.
#Dado un conjunto de puntos se trata de encontrar los dos puntos mas cercanos.
#En 1D. Dada la lista [3403,4537,9089,9746,7259] crear un programa en python para encontrar los dos puntos mas cercanos usando fuerza bruta.
#Como es un problema de fuerza bruta (O(n²)), simplemente comparamos todas las parejas posibles y nos quedamos con la mínima distancia absoluta.
def puntos_mas_cercanos(lista):
    n = len(lista)

    # Inicializamos con un par cualquiera
    min_dist = float('inf')
    par_min = (None, None)

    # Fuerza bruta: todas las combinaciones
    for i in range(n):
        for j in range(i + 1, n):
            dist = abs(lista[i] - lista[j])

            if dist < min_dist:
                min_dist = dist
                par_min = (lista[i], lista[j])

    return par_min, min_dist


# Lista dada
numeros = [3403, 4537, 9089, 9746, 7259]

par, distancia = puntos_mas_cercanos(numeros)

print("Par más cercano:", par)
print("Distancia mínima:", distancia)
```

```
#Dada la lista [3403,4537,9089,9746,7259] crear un programa en python para encontrar los dos puntos mas cercanos aplicando divide y venceras.
#Idea del enfoque Divide y Vencerás
#Ordenamos la lista
#Dividimos en dos mitades
#Resolvemos recursivamente en cada mitad
#Comparamos:
#mejor distancia izquierda
#mejor distancia derecha
#frontera (split)
def distancia(p1, p2):
    return abs(p1 - p2)


def fuerza_bruta(arr):
    min_dist = float('inf')
    par = (None, None)

    for i in range(len(arr)):
        for j in range(i + 1, len(arr)):
            d = distancia(arr[i], arr[j])
            if d < min_dist:
                min_dist = d
                par = (arr[i], arr[j])

    return par, min_dist


def divide_y_venceras(arr):
    # Caso base
    if len(arr) <= 3:
        return fuerza_bruta(arr)

    mid = len(arr) // 2
    izquierda = arr[:mid]
    derecha = arr[mid:]

    par_izq, dist_izq = divide_y_venceras(izquierda)
    par_der, dist_der = divide_y_venceras(derecha)

    # Mejor de los dos lados
    if dist_izq < dist_der:
        mejor_par = par_izq
        mejor_dist = dist_izq
    else:
        mejor_par = par_der
        mejor_dist = dist_der

    # Comparación cruzada (frontera)
    for i in izquierda:
        for j in derecha:
            d = distancia(i, j)
            if d < mejor_dist:
                mejor_dist = d
                mejor_par = (i, j)

    return mejor_par, mejor_dist


# Lista dada
nums = [3403, 4537, 9089, 9746, 7259]

# IMPORTANTE: se ordena antes
nums.sort()

par, dist = divide_y_venceras(nums)

print("Par más cercano:", par)
print("Distancia mínima:", dist)
```

```
#2D dada la matriz [[1122,6175],[135,4076],[7296,2741]] crear un programa en python que devuelva los dos puntos mas cercanos
#Este es el problema clásico de “Closest Pair of Points” en 2D,
#y la forma eficiente de resolverlo es con Divide y Vencerás (O(n log n)).
#Idea del algoritmo
#Ordenar los puntos por coordenada x
#Dividir en dos mitades
#Resolver recursivamente:
#mejor par izquierda
#mejor par derecha
#Combinar:
#comparar pares cercanos a la “frontera” (strip)
#Calcular distancia euclídea
#Fórmula de distancia
#d=(x1​−x2​)2+(y1​−y2​)2
import math


def distancia(p1, p2):
    return math.sqrt((p1[0] - p2[0])**2 + (p1[1] - p2[1])**2)


def fuerza_bruta(puntos):
    min_dist = float('inf')
    par = (None, None)

    for i in range(len(puntos)):
        for j in range(i + 1, len(puntos)):
            d = distancia(puntos[i], puntos[j])
            if d < min_dist:
                min_dist = d
                par = (puntos[i], puntos[j])

    return par, min_dist


def strip_closest(strip, d_min):
    min_dist = d_min
    par = None

    strip.sort(key=lambda p: p[1])  # ordenar por y

    for i in range(len(strip)):
        for j in range(i + 1, len(strip)):
            if (strip[j][1] - strip[i][1]) >= min_dist:
                break

            d = distancia(strip[i], strip[j])
            if d < min_dist:
                min_dist = d
                par = (strip[i], strip[j])

    return par, min_dist


def divide_y_venceras(puntos_sorted_x):
    n = len(puntos_sorted_x)

    if n <= 3:
        return fuerza_bruta(puntos_sorted_x)

    mid = n // 2
    mid_point = puntos_sorted_x[mid]

    left = puntos_sorted_x[:mid]
    right = puntos_sorted_x[mid:]

    (p1, d1) = divide_y_venceras(left)
    (p2, d2) = divide_y_venceras(right)

    if d1 < d2:
        d_min = d1
        best_pair = p1
    else:
        d_min = d2
        best_pair = p2

    # construir strip (frontera)
    strip = [
        p for p in puntos_sorted_x
        if abs(p[0] - mid_point[0]) < d_min
    ]

    p_strip, d_strip = strip_closest(strip, d_min)

    if p_strip and d_strip < d_min:
        return p_strip, d_strip

    return best_pair, d_min


# -------------------------
# Datos del problema
# -------------------------
puntos = [[1122, 6175], [135, 4076], [7296, 2741]]

# ordenar por x
puntos.sort(key=lambda p: p[0])

par, dist = divide_y_venceras(puntos)

print("Par más cercano:", par)
print("Distancia mínima:", dist)
```
