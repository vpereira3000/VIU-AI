RELACION DE CORRELACION (η) → CATEGORICA VS NUMERICA

La razón de correlación (η, Eta) es el estadístico adecuado para medir la fuerza de la asociación entre una variable categórica y una variable numérica.

A diferencia de la correlación de Pearson, η no presupone una relación lineal y funciona de manera natural con variables nominales, tales como el género, la frecuencia de compra o el método de búsqueda de productos.


El valor oscila entre:

0 → sin asociación
1 → asociación perfecta


Una interpretación común es:
| η         | Interpretation           |
| --------- | -------------------------|
| 0.00–0.10 | Negligible (Despreciable)|
| 0.10–0.30 | Weak (Debil)             |
| 0.30–0.50 | Moderate (Moderada)      |
| >0.50     | Strong (Fuerte)          |


```python
def correlation_ratio(categories, measurements):
    """
    Computes Correlation Ratio (Eta)

    Parameters
    ----------
    categories : categorical variable
    measurements : numerical variable

    Returns
    -------
    eta : float
    """

    # Remove missing values
    valid = pd.notnull(categories) & pd.notnull(measurements)

    categories = np.array(categories[valid])
    measurements = np.array(measurements[valid], dtype=float)

    # Convert categories to integer codes
    codes, uniques = pd.factorize(categories)

    n_categories = len(uniques)

    grand_mean = np.mean(measurements)

    numerator = 0
    denominator = np.sum((measurements - grand_mean) ** 2)

    for i in range(n_categories):

        values = measurements[codes == i]

        if len(values) == 0:
            continue

        mean_i = np.mean(values)

        numerator += len(values) * (mean_i - grand_mean) ** 2

    eta = np.sqrt(numerator / denominator)

    return eta
```

EJEMPLO 1. RELACIÓN ENTRE EL GÉNERO Y LA SATISFACCIÓN CON LAS COMPRAS.

```python
eta = correlation_ratio(
    df["Gender"],
    df["Shopping_Satisfaction"]
)

print("Eta =", round(eta,4))
```

CALCULAR Η PARA TODAS LAS VARIABLES CATEGÓRICAS FRENTE A TODAS LAS NUMÉRICAS.

ESTE ES EL ANÁLISIS QUE HABITUALMENTE SE INCLUYE EN UN ANÁLISIS EXPLORATORIO DE DATOS (EDA) PROFESIONAL.

```python
categorical = df.select_dtypes(include="object").columns
numerical = df.select_dtypes(include="number").columns

eta_matrix = pd.DataFrame(
    index=categorical,
    columns=numerical
)

for cat in categorical:

    for num in numerical:

        eta_matrix.loc[cat, num] = correlation_ratio(
            df[cat],
            df[num]
        )

eta_matrix = eta_matrix.astype(float)

print(eta_matrix)
```

```python
plt.figure(figsize=(9,8))

plt.imshow(
    eta_matrix,
    cmap="YlOrRd",
    aspect="auto",
    vmin=0,
    vmax=1
)

plt.colorbar(label="Correlation Ratio (η)")

plt.xticks(
    range(len(eta_matrix.columns)),
    eta_matrix.columns,
    rotation=45,
    ha="right"
)

plt.yticks(
    range(len(eta_matrix.index)),
    eta_matrix.index
)

# Show values
for i in range(len(eta_matrix.index)):
    for j in range(len(eta_matrix.columns)):
        plt.text(
            j,
            i,
            f"{eta_matrix.iloc[i,j]:.2f}",
            ha="center",
            va="center",
            fontsize=8
        )

plt.title("Correlation Ratio (η)\nCategorical vs Numerical Variables")

plt.tight_layout()
plt.show()
```

Interpretación de la matriz de razón de correlación (η)

La matriz de razón de correlación (η) ofrece información valiosa sobre la intensidad de la asociación entre las características categóricas de los clientes y los indicadores numéricos de comportamiento. A diferencia de la correlación de Pearson, η captura en qué medida la pertenencia a una categoría específica explica la variabilidad de una variable numérica, sin asumir una relación lineal.

De esta visualización se pueden extraer varias conclusiones prácticas:

1 Identificar las características del cliente más influyentes

Los valores de η más elevados indican las variables categóricas que ejercen una mayor influencia sobre los resultados numéricos. Por ejemplo, si la frecuencia de compra presenta un valor de η alto en relación con la satisfacción de compra, esto sugiere que los hábitos de compra están fuertemente asociados a los niveles de satisfacción del cliente.

Desde una perspectiva empresarial, estas variables merecen especial atención, ya que representan posibles factores determinantes del comportamiento del cliente.

2 Detectar variables con escaso impacto práctico

Las variables con valores η cercanos a cero contribuyen muy poco a explicar la variación en las variables numéricas.

Por ejemplo, si la variable Género presenta consistentemente valores η inferiores a 0,10 en todas las medidas numéricas, esto sugiere que el comportamiento del cliente es en gran medida independiente del género en este conjunto de datos.

Esta información es igualmente valiosa porque permite a los analistas eliminar variables que probablemente no mejoren los modelos predictivos.

3 Priorizar variables para modelos predictivos

La matriz η sirve como una herramienta eficaz para la selección de variables.

Las variables categóricas que presentan asociaciones moderadas o fuertes (η > 0,30) son buenas candidatas para su inclusión en:

Segmentación de clientes
Sistemas de recomendación
Predicción de la satisfacción
Pronóstico del comportamiento de compra
Modelos de clasificación de aprendizaje automático

Por el contrario, las variables con valores de η consistentemente insignificantes pueden aportar poco valor predictivo.

4 Comprender el comportamiento del cliente

Una de las principales fortalezas de la matriz η es que revela qué características de los clientes explican las diferencias de comportamiento.

Por ejemplo, unas asociaciones sólidas entre

la frecuencia de navegación,
la frecuencia de compra y
la frecuencia de recomendación

y

la satisfacción con la compra y
la precisión de las valoraciones

sugerirían que los clientes con un alto nivel de vinculación tienden a valorar los servicios de Amazon de manera diferente a los compradores ocasionales.

Esta información puede orientar las estrategias de marketing dirigidas a aumentar la vinculación de los clientes.

5 Apoyo a la toma de decisiones empresariales

Desde una perspectiva gerencial, la matriz η ayuda a identificar qué características del cliente merecen inversión.

Si las variables relacionadas con las recomendaciones muestran una fuerte asociación con la satisfacción, Amazon podría priorizar la mejora de los algoritmos de recomendación.

Si la importancia de las reseñas está altamente asociada con el comportamiento de compra, los esfuerzos podrían centrarse en incentivar a los clientes a publicar reseñas de productos.

Estos hallazgos permiten a quienes toman las decisiones concentrar los recursos en los factores con mayor impacto medible.

6 Mejora la segmentación de clientes

Las variables con valores altos de η distinguen naturalmente a los grupos de clientes.

Estas variables son excelentes candidatas para algoritmos de agrupación (clustering) como:

K-Means
Agrupación jerárquica
DBSCAN

ya que representan diferencias de comportamiento significativas entre los clientes.

7 Orientar futuras investigaciones

La matriz η identifica relaciones que merecen un análisis estadístico más profundo.

Las variables que presentan asociaciones fuertes pueden analizarse posteriormente mediante:

ANOVA
Modelos de regresión
Árboles de decisión

para cuantificar con mayor precisión su contribución predictiva.

**CONCLUSION GENERAL**

La matriz de razón de correlación (η) es una herramienta poderosa de análisis exploratorio para datos de encuestas mixtos, ya que cuantifica en qué medida las características categóricas de los clientes explican la variación en las medidas numéricas de comportamiento. Más que identificar relaciones de causa y efecto, pone de relieve asociaciones significativas que pueden orientar análisis posteriores y decisiones empresariales.

En el contexto de la Encuesta sobre el Comportamiento del Cliente de Amazon, la matriz η permite a los analistas identificar los atributos del cliente más estrechamente asociados con la satisfacción, el comportamiento de compra y la eficacia de las recomendaciones. Estos hallazgos pueden guiar la selección de variables para modelos predictivos, mejorar las estrategias de segmentación de clientes y respaldar decisiones basadas en datos orientadas a optimizar la experiencia del cliente. Al mismo tiempo, es posible restar prioridad a las variables con valores de η consistentemente bajos, simplificando así los análisis futuros sin sacrificar la capacidad explicativa.

En conjunto, el análisis de la razón de correlación complementa la correlación de Pearson y la V de Cramér, proporcionando una comprensión integral de las relaciones entre variables de distinto tipo en los conjuntos de datos sobre el comportamiento del cliente.
